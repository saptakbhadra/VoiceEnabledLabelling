```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
import speech_recognition as sr
import os
```

```python
def take_speech():
    p=sr.Recognizer()
    with sr.Microphone() as source:
        print("Talk")
        audio_to_text=p.listen(source)
        print('Done')
        try:
            speech2text=p.recognize_google(audio_to_text)
            return (speech2text)
        except:
            return("Try Again !")
```

```python
directory='D:/PythonFiles/Projects/Labelling/images/'
for filename in os.listdir(directory):
    if filename.endswith((".jpg",".png",".jpeg")):
        print(filename)
        img=Image.open('D:/PythonFiles/Projects/Labelling/images/'+filename)
        plt.imshow(img)
        plt.show()
        os.rename('D:/PythonFiles/Projects/Labelling/images/'+filename,'D:/PythonFiles/Projects/Labelling/images_labelled/'+'{}.jpg'.format(take_speech()))
        print(filename)
        continue
    else:
        continue
```
